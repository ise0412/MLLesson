````python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df_flights = sns.load_dataset('flights')
df_flights.columns=['年','月','人数']
df_flights.head(5)
#df = DataFrame(np.arange(0, 12).reshape(6, -1), columns = ['始め','終わり'])
````

````python
df_flights_pivot = pd.pivot_table(data=df_flights, values='人数', 
                                  columns='年', index='月', aggfunc=np.mean);
df_flights_pivot.plot();
````

````python
font_paths = matplotlib.font_manager.findSystemFonts()
font_objects = matplotlib.font_manager.createFontList(font_paths)
font_name = [f.name for f in font_objects]

mincho = [name for name in font_name if 'mincho' in name.lower()]
gothic = [name for name in font_name if 'gothic' in name.lower()]
print("明朝系フォント",mincho)
print("ゴシック系フォント",gothic)
````

````python
plt.rcParams['font.family'] = '07YasashisaGothic' #全体のフォントを設定
````

````python
df_flights_pivot.plot();
````

# フォントのダウンロード

[IPAexフォント](http://ipafont.ipa.go.jp/)から IPAex2書体パックをダウンロードする

# フォントのインストール

## 管理者になれる場合

管理者になれる場合は、ダウンロードしたファイルをダブルクリックで開き、中にある2つのフォント、 
- ipaexg(.ttf)  ゴシック
- ipaexm(.ttf)　明朝
をダブルクリックしてシステムにインストールします。

## 管理者になれない場合

### 1. IPAexフォントをユーザ配下の .matplotlib\fonts\ttf にコピー

**.matplotlib\fonts\ttf**  というフォルダがなければ作る

C:\Users\User名\.matplotlib\fonts\ttf  

### 2. matplotlibrc の場所を確認
```
import matplotlib
print(matplotlib.matplotlib_fname)
```
### 3. matplotlibrc をユーザ配下の .matplotlib にコピー

デフォルトでは、C:\ProgramData\Anaconda3\lib\site-packages\matplotlib\mpl-data\matplotlibrc である。

### 4. デフォルトフォント設定

コピー出来たらメモ帳やAtomエディタなどのテキストエディタで開き、

** font.family : IPAexGothic **

という行を追加する。






````python
import matplotlib
print(matplotlib.matplotlib_fname())
````

````python
plt.rcParams['font.family'] = 'Tanuki Permanent Marker' #全体のフォントを設定

plt.rcParams["figure.figsize"] = [12, 10]
plt.rcParams['font.size'] = 20 #フォントサイズを設定 default : 12
plt.rcParams['xtick.labelsize'] = 15 # 横軸のフォントサイズ
plt.rcParams['ytick.labelsize'] = 15


df_flights_pivot.plot();
````
